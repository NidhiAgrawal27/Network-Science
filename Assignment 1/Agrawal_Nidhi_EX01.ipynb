```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
                          # Method to calculate average degree of the network

def avgdegree(graph):
    degrees = list(graph.degree(i) for i in graph.nodes())   # Creating list of degrees of all nodes of the graph
    k = sum(degrees)/len(degrees)                            # Calculating average degree
    print("Average degree of the network is" , 
          round(k,3), '\n')                                  # Print average degree rounded to 3 decimal places
    return
```

```python
                             # Method to calculate density of the network

def densityfunc(graph):
    edges_num = nx.number_of_edges(graph)               # Fetching total number of edges of the network
    nodes_num = nx.number_of_nodes(graph)               # Fetching total number of nodes of the network
    density = (2*(edges_num))/(nodes_num*(nodes_num-1)) # Calculating density using formula d = 2L/(N(N-1))
    print("Density of the network is", 
          round(density,3), '\n')                       # Print density of the graph rounded to 3 decimal places
    return

#nx.density(graph)
```

```python
                    # Method to fetch maximum degree and list of nodes with maximum degree

def max_degree(graph):
    maxitem = max(graph.degree(), 
                  key=lambda x: x[1])   # Fetch tuple (node,degree) with max degree from dictionary graph.degree()
    maxdegree = maxitem[1]              # Maximum degree
    maxnode = maxitem[0]                # First node with maximum degree
    
    
    # In case of multiple nodes with same maximum degree value: 
    
    maxnodeslist = list()               # Creatye an empty list
    for node,degree in graph.degree():  # Loop to create a list of all nodes with maximum degree 
        if degree == maxdegree:
            maxnodeslist.append(node)   # Appending all nodes with maximum degree to the list 'nodes'
    return maxnodeslist,maxdegree       # Return a list of nodes with maximum degree and the maximum degree
    
```

# Creating a Star Network <br>

```python
gstar = nx.Graph()                     # Creating an empty graph
```

```python
gstar.add_node(1)                      # Adding a node numbered '1' which will be the center of star network
gstar.nodes()                          # View the nodes
```

```python
        # Loop to add 31 nodes (numbered 2 to 32) and edges between central node '1' and every other node
    
for i in range(2,33): 
    gstar.add_node(i)                  # Add node to the graph
    gstar.add_edge(1,i)                # Add edge between the newly added node and central node 1
gstar.edges()                          # View the edges
```

```python

plt.figure(figsize=(7,7))                            # Setting the graph size for better visual
nx.draw(gstar,node_size=500,with_labels=True)        # Plot the graph with increased node size for better visual

plt.title("Star Network",fontsize=22)                # Set title of the graph
plt.show()                                           # Display star network
```

```python
gstar.degree()                              # View degree of the star network
```

```python
avgdegree(gstar)                          # calling method to compute and print average degree of the network
densityfunc(gstar)                        # calling method to compute and print density of the network
nodesList, maxdeg = max_degree(gstar)     # calling method to return list of nodes with maximum degree value and
                                          # maximum degree of the network

print("Maximum degree of the network is:")
print(maxdeg, "\n\nList of nodes with this maximum degree is:")   # print maximum degree
print(*nodesList, sep = ",")                                      # print list of nodes with max degree
```

## Features of the star network:
### Average degree = 1.938
### Density = 0.062
### Maximum degree = 31
### Node with id 1 (which is the central node) has maximum degree of 31 <br><br>

```python
degree_gstar = [gstar.degree(n) for n in gstar.nodes()]  # Creating a list of degrees of each node
plt.hist(degree_gstar, bins = range(0,35))               # Plot a histogram with 35 bins
plt.xlabel('Degree')                                     # Set label of x axis
plt.ylabel('Number of nodes')                            # Set label of y axis
_=plt.title("Degree histogram of Star Network")          # Set title of the graph
```

### From the graph above, we see that there is one node with degree 31 and there are 31 nodes with degree 1 <br><br>

```python
plt.hist(degree_gstar, bins = range(0,35), density = True)    # Plot the probability distribution with 35 bins
plt.xlabel('Degree d')                                        # Set label of x axis
plt.ylabel('Probability a node will have degree d')           # Set label of y axis
plt.title("Degree Distribution")                              # Set title of the graph
plt.show()                                                    # Display graph
```

### Probability a node will have degree 1 = 31/32 = 0.96875
### Probability a node will have degree 31 = 1/32 = 0.03125

### Which is the result in the graph above <br><br>

# Creating a fully connected network <br>

```python
gconn = nx.Graph()                           # Creating an empty graph
```

```python
gconn.add_nodes_from(range(1,33))            # Adding 32 nodes to the graph gconn

for i in range(1,33):                        # Loop to add edges
    for j in range(1,33):
        if i!=j:                             # The node should not be connected by an edge to itself
            gconn.add_edge(i,j)              # Add edges from one node to all other nodes except itself
```

```python
plt.figure(figsize=(10,10))                        # Increasing the size of the graph for better visual
nx.draw_circular(gconn, node_size=500, 
                 with_labels = True)               # Draw graph with bigger node size and nodes placed in a circle
plt.title("Fully Connected Network",fontsize=22)   # Set title of the graph
plt.show()                                         # Display graph
```

```python
avgdegree(gconn)                         # calling method to compute and print average degree of the network
densityfunc(gconn)                       # calling method to compute and print density of the network
nodesList, maxdeg = max_degree(gconn)    # calling method to return list of nodes with maximum degree value and 
                                         # maximum degree of the network

print("Maximum degree of the network is:")
print(maxdeg,"\n\nList of nodes with this maximum degree is:")  # print maximum degree
print(*nodesList, sep = ",")                                    # print list of nodes with maximum degree
```

## Features of the fully connected network:
### Average degree = 31
### Density = 1
### Maximum degree = 31
### All 32 nodes have degree 31 which is also the maximum degree<br><br>

```python
degree_gconn = [gconn.degree(n) for n in gconn.nodes()]        # creating a list of degrees of each node
plt.hist(degree_gconn, bins = range(0,35))                     # Plot a histogram with 35 bins
plt.xlabel('Degree')                                           # Set label for x axis
plt.ylabel('Number of nodes')                                  # Set label for y axis
_=plt.title("Degree histogram of fully connected network")     # Set title for the graph
```

### From the above graph we can see that there are 32 nodes each with degree 31<br>

```python
plt.hist(degree_gconn, bins = range(0,35), density = True)      # Plot probability distribution with 35 bins
plt.xlabel('Degree d')                                          # Set label for x axis
plt.ylabel('Probability a node will have degree d')             # Set label for y axis
_=plt.title("Degree Distribution")                              # Set title for the graph 
```

### Probability a node will  have degree 31 = 32/32 = 1
### Which is the result in the graph above <br><br>

# Creating 1-D lattice with degree 2 for each node <br>

```python
g = nx.Graph()                              # Create an empty graph
```

```python
g.add_nodes_from(range(1,33))               # Add 32 nodes to graph

for y in range(1,32):                       # Loop to add edges to the nodes
    g.add_edge(y,y+1)                       # Add edge from one node to it's next node

    
g.add_edge(1,y+1)                           #Add edge between first and last node
    
```

```python
plt.figure(figsize=(7,7))                               # Set graph size for better visual
nx.draw_circular(g ,node_size=500, with_labels=True)    # Draw graph in circular shape with bigger node size
plt.title("1-D lattice with degree 2 for each node",
          fontsize=22)                                  # Set title of the graph    
plt.show()                                              # Display graph
```

```python
avgdegree(g)                             # calling method to compute and print average degree of the network
densityfunc(g)                           # calling method to compute and print density of the network
nodesList, maxdeg = max_degree(g)        # calling method to return list of nodes with maximum degree value and 
                                         # maximum degree of the network

print("Maximum degree of the network is:")
print(maxdeg, "\n\nList of nodes with this maximum degree is:") # print maximum degree of the network
print(*nodesList, sep = ",")                                    # print list of nodes with maximum degree 
```

# Creating network for dataset graph_facebook.gml<br>

```python
g_fb = nx.read_gml("OneDrive_1_9-21-2021/graph_facebook.gml")    # Reading given dataset
plt.figure(figsize=(12,12))                                      # Increase figure size for better visual
nx.draw(g_fb, with_labels=True)                                  # Draw graph for given dataset
plt.title("Facebook Network\nEdges: Friendship relation between users\
\nNodes: Facebook users", fontsize=20)                           # Set title of the graph
plt.show()                                                       # Display graph
```

```python
avgdegree(g_fb)                          # calling method to compute and print average degree of the network
densityfunc(g_fb)                        # calling method to compute and print density of the network
nodesList, maxdeg = max_degree(g_fb)     # calling method to return list of nodes with maximum degree value and 
                                         # maximum degree of the network

print("Maximum number of friends a user has on Facebook is:")
print(maxdeg, "\n\nList of users' id number with maximum number of friends is:")  # print maximum degree
print(*nodesList, sep = ",")                                         # print list of nodes with maximum degree
```

## Features of the Facebook network:
### Average degree = 43.691
### Density = 0.011
### Maximum degree = 1045 (this is the maximum number of friends a user has as per Facebook dataset)
### Node with id 107 has maximum degree of 1045 (this means that the user id 107 has maximum number of friends (1045 friends) on facebook)<br><br>

```python
degree_g_fb = [g_fb.degree(n) for n in g_fb.nodes()]    # creating a list of degrees of each node
plt.hist(degree_g_fb, bins = range(0,1200,50), 
         edgecolor='Black')                             # Plot a histogram                           
plt.xlabel('Number of friends')                         # Set label for x axis 
                                                        # Edges represent friendship relation between users
    
plt.ylabel('Number of Facebook users')                  # Set label for y axis 
                                                        # Nodes represent Facebook users for this dataset
    
_=plt.title("Degree histogram for Facebook")            # Set title of the graph
```

### We can see that the maximum degree of 1045 is not visible on the x axis. We will now plot probability distribution and see the result<br>

```python
plt.hist(degree_g_fb, bins = range(0, 1200), density = True)          # Plot the histogram
plt.xlabel('Number of friends n')                                     # Set the label for x axis
plt.ylabel('Probability a user will have n number of friends')        # Set the label for y axis
plt.title("Degree Distribution for Facebook")                         # Set the title for the graph
plt.show()                                                            # Display the graph
```

### We can see that the maximum degree of 1045 is now visible on the x axis but extremely small and therefore it is difficult to infer anything from the graph. Also, the data distribution is accumulated in a small space on the left of the graph. Therefore to better view and understand the distribution, we will plot the graph using log scale.<br>

```python
plt.figure(figsize=(8,8))
bins_num = np.logspace(np.log10(min(degree_g_fb)),
                      np.log10(max(degree_g_fb)+1),30)     # Set number of bins for histogram as per logscale
_ = plt.hist(degree_g_fb, bins = bins_num,
             density = True, edgecolor='Black')             # Plot a probability distribution using histogram
plt.xscale('log')                                           # set log scale for x axis
plt.yscale('log')                                           # set log scale for y axis
plt.xlabel('Number of friends n',fontsize=15)                              # set label for x axis
plt.ylabel('Probability a user will have n number of friends',fontsize=15) # set label for x axis
_=plt.title("Degree Distribution for Facebook",fontsize=15)                # set title of the graph
```

### The graph above is much easier to infer as we can visualize the distribution better in logscale. 
### We can deduce that the probability of a user having more than 500 friends becomes miniscule and the least probability is of having more than 1000 friends. <br><br>
