```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python

# Method to plot average degree of the nearest neighbours knn(k) as a function of the vertices degree k 
# for both real and randomized network in the same plot

def plotAvgDegreeNbr_realAndRand(realgraph, randomgraph, filename):
    
    # Get a dictionary of degree and average neighbour degree for real network
    avg_nbr_deg_real = nx.average_degree_connectivity(realgraph)
    
    # Get a dictionary of degree and average neighbour degree for randomized network
    avg_nbr_deg_random = nx.average_degree_connectivity(randomgraph)
    
    # Get degrees in a list 'deg_real' and corresponding average neighbour degrees in list 'avgNbrDeg_real' 
    # for real network
    deg_real, avgNbrDeg_real = zip(*avg_nbr_deg_real.items())
    
    # Get degrees in a list 'deg_rnd' and corresponding average neighbour degrees in list 'avgNbrDeg_rnd' 
    # for randomized network
    deg_rnd, avgNbrDeg_rnd = zip(*avg_nbr_deg_random.items())
    
    # Set graph size for better visual
    plt.figure(figsize=(8,8))
    
    # Plot average degree of the nearest neighbours knn(k) versus vertices degree k
    plt.plot(deg_real,avgNbrDeg_real,'o')                 # Scatter plot for real graph
    plt.plot(deg_rnd,avgNbrDeg_rnd,'o',color='red')       # Scatter plot for randomized graph
    plt.legend(["Real Network","Randomized Network"])     # Set label for data that is plotted
    plt.xlabel("Degree k", fontsize = 15)                                # Set label for x axis
    plt.ylabel("Average degree of neighbour knn(k)", fontsize = 15)      # Set label for y axis
    plt.title(filename, fontsize = 15, loc = 'left')                     # Set title of the graph
    plt.show()                                                           # Display graph
    
```

```python

                        # Method to calculate average degree of the network

def avgDegree(graph):
    degrees = list(graph.degree(i) for i in graph.nodes())   # Creating list of degrees of all nodes of the graph
    k = sum(degrees)/len(degrees)                            # Calculating average degree
    return k

```

```python

                # Method to plot degree distribution of the network in double log scale 

def plotDegreeDist(graph,filename,label):
    degree = [graph.degree(n) for n in graph.nodes()]          # creating a list of degrees of each node
    bins_num = np.logspace(np.log10(min(degree)),
                      np.log10(max(degree)+1),30)              # Set number of bins for histogram as per logscale
    plt.figure(figsize=(8,8))                                  # Set graph size for better visual
    plt.hist(degree, bins = bins_num, density = True, edgecolor='Black') # Plot the probability distribution
    plt.xlabel(label, fontsize = 15)                                               # Set label of x axis
    plt.ylabel('Probability of degree k', fontsize = 15)                           # Set label of y axis
    plt.title("Degree Distribution of " + filename, fontsize = 15, loc = 'left')   # Set title of the graph
    plt.xscale('log')                                                  # set log scale for x axis
    plt.yscale('log')                                                  # set log scale for y axis
    k = avgDegree(graph)                               # call method to calculate the average degree of the graph
    plt.axvline(k, color = 'Red')                      # plot a vertical line for average degree
    plt.legend(["Average degree = "+ str(round(k,2))]) # Set label for vertical line (average degree)
    plt.show()                                         # Display graph

    
```

```python

# Method for Scatter plot of clustering coefficient of all nodes in real and randomized network: ùê∂(ùëñ) vs. ùê∂ùëüùëõùëë(ùëñ)
# Clustering coefficient in the real graph ùí¢ (denoted by ùê∂(ùëñ)) and randomized graph ùí¢ùëüùëõùëë (denoted as ùê∂ùëüùëõùëë(ùëñ))


def clusteringCoeff(realgraph, randomgraph, filename):
    clust_real = nx.clustering(realgraph)            # compute clustering coefficient of nodes of real graph
    c_i = list(clust_real.get(i)
               for i in realgraph.nodes())           # list of clustering coefficients of real graph
    
    clust_random = nx.clustering(randomgraph)        # compute clustering coefficient of nodes of randomized graph
    c_rnd = list(clust_random.get(i) 
                 for i in randomgraph.nodes())       # list of clustering coefficients of randomized graph
    
    plt.figure(figsize=(8,8))                                   # Set graph size for better visual    
    plt.plot(c_i,c_rnd,'o')                                     # Scatter plot c_i Versus c_rnd
    plt.ylabel("Clustering coefficient of randomized network", fontsize = 15)  # Set label for y axis
    plt.xlabel("Clustering coefficient of real network", fontsize = 15)        # Set label for x axis
    plt.title(filename, fontsize = 15, loc = 'left')                           # Set title of the graph
    plt.show()                                                                 # Display graph
    
```

```python

filelist = os.listdir("data")   # List of all file names containing datasets in directory data

# Loop to read all files in filelist and explore their properties
for file in filelist:
    
    # Loop to set name of the networks and the edges for use in title and label of the graphs
    if file == "graph_AstroPh.gml" :
        filename = 'Astro Physics Arxiv Collaborations Network'
        edgesName = 'No. of co-authourships (k)'
        
    elif file == "graph_jazz_collab.gml" :
        filename = 'Jazz Collaboration Network'
        edgesName = 'No. of collaborations in bands (k)'
        
    elif file == "graph_internet.gml" :
        filename = 'Internet AS Network'
        edgesName = 'No. of connections between autonomous systems (k)'
        
    elif file == "graph_game_thrones.gml" :
        filename = 'Game of Thrones Coappearances Network'
        edgesName = 'No. of co-appearances of characters (k)'
        
    elif file == "graph_dolphins.gml" :
        filename = 'Dolphin Social Network'
        edgesName = 'No. of frequent associations (k)'
        
    elif file == "graph_florentine_families.gml" :
        filename = 'Florentine Families Network'
        edgesName = 'No. of marriage alliances & business relationships (k)'
        
    elif file == "graph_terrorists.gml" :
        filename = '9-11 Terrorist Network'
        edgesName = 'No. of known social associations (k)'
        
    elif file == "graph_eu_airlines.gml" :
        filename = 'European Airline Network'
        edgesName = 'No. of airline routes (k)'
        
    elif file == "graph_celegansInteractomes.gml" :
        filename = 'Caenorhabditis Elegans Interactomes Network'
        edgesName = 'No. of protein-protein interactions (k)'
        
    elif file == "graph_facebook.gml" :
        filename = 'Facebook Friendships Network'
        edgesName = 'No. of friendship relations (k)'
        
    elif file == "graph_CondMat.gml" :
        filename = 'Condensed Matter Arxiv Collaborations Network'
        edgesName = 'No. of co-authourships (k)'
        
    elif file == "graph_chess.gml" :
        filename = 'Kaggle Chess Players Network'
        edgesName = 'No. of chess matches (k)'
        
    
    # Read dataset
    graph = nx.read_gml(os.path.join("data/", file))
    
    real_filename = filename + " - Real Network"      # Set Real network name for use in title of the graphs
    rnd_filename = filename + " - Randomized Network" # Set Randomized network name for use in title of the graphs
    
    print("\n\n\n\n\n**************** DATASET : " + filename + " ****************\n\n\n\n\n")
    
    # Randomize the network
    grnd_graph = nx.algorithms.smallworld.random_reference(graph, niter = 10, connectivity=False)
    
    print("\n\n")
    
    # call method to plot average degree of neighbour versus degree of both real and randomized network
    plotAvgDegreeNbr_realAndRand(graph ,grnd_graph, filename)
    
    print("\n\n")
    
    # Assortativity coefficient of the real network
    a_real = nx.degree_assortativity_coefficient(graph)
    print("\n\nAssortativity coefficient of real " + filename + " is " +  str(round(a_real,3)) + "\n")
    
    # Assortativity coefficient of the randomized network
    a_rnd = nx.degree_assortativity_coefficient(grnd_graph)
    print("Assortativity coefficient of randomized  " + filename + " is " +  str(round(a_rnd,3)) + "\n\n")
    
    print("\n\n")
    
    # Call method 'plotDegreeDist' to plot degree distribution of the real network in double log scale
    plotDegreeDist(graph, real_filename, edgesName)
    
    print("\n\n")
    
    # Call method 'plotDegreeDist' to plot degree distribution of the randomized network in double log scale
    plotDegreeDist(grnd_graph, rnd_filename, edgesName)
    
    print("\n\n")
    
    # Call method 'clusteringCoeff' to create a scatter plot of clustering coefficient of all nodes in real and randomized network
    clusteringCoeff(graph, grnd_graph, filename)
    
    print("\n\n")
    
    
```
